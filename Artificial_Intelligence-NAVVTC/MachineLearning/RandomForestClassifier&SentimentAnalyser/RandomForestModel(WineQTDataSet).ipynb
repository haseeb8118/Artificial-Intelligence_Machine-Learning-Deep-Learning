## Data processing and Test-data spilit

```python
msk = np.random.rand(len(dt)) < 0.08
train = dt[msk]
test = dt[~msk]
```

```python
from sklearn.model_selection import train_test_split
X = dt.drop(columns = 'quality')
y = dt['quality']
```

## RANDOM FOREST MODEL

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
print('Accuracy: ', metrics.accuracy_score(y_test,y_pred))
```

```python
dt.columns
```

```python
import pandas as pd
feature_imp = pd.Series(clf.feature_importances_, index=dt.columns[:12]).sort_values(ascending=False)
feature_imp
```

## Importance Data Visualization

```python
%matplotlib inline
import seaborn as sns

sns.barplot(x=feature_imp, y=feature_imp.index)

plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.legend()
plt.show()
```
