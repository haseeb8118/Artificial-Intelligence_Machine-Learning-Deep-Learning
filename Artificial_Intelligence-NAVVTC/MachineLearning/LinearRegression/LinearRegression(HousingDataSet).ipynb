# Data Spiliting

```python
msk = np.random.rand(len(dataa)) < 0.08
train = dataa[msk]
test = dataa[~msk]
```

# Model Selection and training

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.33, random_state = 1000)
```

# Model Evaluation

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x_train = np.asanyarray(train[['total_rooms']])
y_train = np.asanyarray(train[['median_house_value']])
regr.fit(x_train, y_train)
print("Coefficients:",regr.coef_)
print("Intercept:",regr.intercept_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(train['total_rooms'], train['median_house_value'], color='blue')
plt.plot(x_train,regr.coef_[0][0]* x_train + regr.intercept_[0],'r')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=  LogisticRegression()
logmodel.fit(x_train, y_train)
```

```python
predictions = logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```
