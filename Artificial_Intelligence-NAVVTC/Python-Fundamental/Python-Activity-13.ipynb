Q1.Create a line plot using Matplotlib to visualize a series of data points.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 25, 18, 30, 12]

plt.plot(x, y, marker='o', linestyle='-', color='b', label='Data Series')

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot Example')

plt.grid(True)
plt.show()
```

Q2.Create multiple subplots within a single figure using Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].plot(x, y1, color='b')
axs[0, 0].set_title('Subplot 1: sin(x)')

axs[0, 1].plot(x, y2, color='g')
axs[0, 1].set_title('Subplot 2: cos(x)')

axs[1, 0].plot(x, y3, color='r')
axs[1, 0].set_title('Subplot 3: tan(x)')

axs[1, 1].plot(x, y4, color='purple')
axs[1, 1].set_title('Subplot 4: exp(x)')

plt.tight_layout()

plt.show()
```

Q3.Create a bar chart using Matplotlib to compare different categories. 

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 50]

plt.bar(categories, values, color='blue')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

Q4.Create a scatter plot using Matplotlib to visualize the relationship between two variables.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)  # X-axis values (random data)
y = 2 * x + 1 + np.random.rand(50)  # Y-axis values (example linear relationship)

plt.scatter(x, y, color='blue', marker='o', label='Data Points')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

plt.legend()

plt.grid(True)
plt.show()
```

Q5.Create a histogram using Matplotlib to visualize the distribution of data.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=20, color='blue', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

plt.grid(True)
plt.show()
```
